```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("Water.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
# Convert 'date' column to datetime format
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Extract only the year
df["year"] = df["date"].dt.year
```

```python
## Calculating KPIs
```

```python
df["water_production_value"] = pd.to_numeric(df["water_production_value"], errors="coerce")
df["water_consumption_value"] = pd.to_numeric(df["water_consumption_value"], errors="coerce")
df["population('000)"] = pd.to_numeric(df["population('000)"], errors="coerce")
```

```python
# 1. NRW %
df.loc[df["sector"] == "overall", "NRW_percent"] = ((df["water_production_value"] - df["water_consumption_value"])
     / df["water_production_value"]) * 100

# 2. Stress Index
df.loc[df["sector"] == "overall", "Stress_Index"] = (df["water_consumption_value"] / df["water_production_value"])

# 3. LPCD
df.loc[df["sector"] == "overall", "LPCD"] = ((df["water_consumption_value"] * 1000)
    / (df["population('000)"] ))
```
